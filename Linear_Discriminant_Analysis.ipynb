```python
#Amaresh Muddebihal 20030141IT015
#Linear Discriminant Analysis

import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
file_path = '/content/iris.data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
print("Initial Data:")
print(data.head())

# Separate features and target variable
X = data.iloc[:, :-1]  # Features - assuming the target is the last column
y = data.iloc[:, -1]   # Target variable

# Convert categorical labels to numerical labels if necessary
le = LabelEncoder()
y = le.fit_transform(y)

# Perform LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

# Plotting the LDA components
plt.figure(figsize=(8, 6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.title('Linear Discriminant Analysis')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.colorbar()
plt.show()
```
